```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Import our balanced dataset

```python
df = pd.read_excel('manifest_balanced.xlsx')
```

```python
def rename_group(name):
    d = {'Arkhangelskie': 'Архангельские',
         'Desninskie': 'Верхне-Деснинские',
         'Donskie': 'Донские',
         'Kostromskie': 'Костромские',
         'Mezhzonalnie': 'Межзональная группа Б',
         'Novgorodskie': 'Новгородские',
         'Povolzkie': 'Владимирско-Поволжские',
         'Pskovskie': 'Псковские',
         'Ryazanskie': 'Рязанские',
         'Seligerskie': 'Селигеро-Торжковские'}
    return d[name]
```

```python
df['variant_rus'] = df['Variant'].apply(rename_group)
```

# Statistics

Distribution of audio segment durations

```python
plt.figure(figsize=(10, 6), dpi=300)
sns.histplot(df, x="duration", color='black')
plt.xlabel('Длительность (сек.)')
plt.ylabel('Количество')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_palette("Greys", 10)
plt.figure(figsize=(10, 6), dpi=300)
sns.boxplot(x=df["duration"], y=df['variant_rus'], linewidth=0.3, fliersize=2, 
            flierprops={"marker": "o", "markerfacecolor":'None'}, 
            order=['Архангельские', 'Костромские', 'Владимирско-Поволжские', 'Новгородские', 'Псковские', 
                   'Селигеро-Торжковские', 'Верхне-Деснинские', 'Межзональная группа Б', 'Рязанские', 'Донские'])
plt.axvline(df["duration"].median(), c='black', ls='--', lw=1)
plt.xlabel('Длительность (сек.)')
plt.ylabel('Группа говоров')
plt.show()
```
